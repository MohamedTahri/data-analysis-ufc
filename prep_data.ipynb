### Preparing data for analysis

```python
# library to easily deal with data
import pandas as pd
```

```python
df = pd.read_csv('data.csv') # importing the data
df.head(2)
```

```python
df = df.set_index('Athlete') # reindexing the dataframe with the list of athletes
df.head(2)
```

```python
df.info()
```

First of all, The dataset contains 960 rows and about 39 data columns. Each row represents a fighter's stats, hometown, reach,...

the columns 'wins by : KO/TKO %', 'wins by : Decision %', 'wins by Submission %' and 'Wins by Knockout','Wins by Submission', 'Wins by Decision' contain the same information. The same thing with 'Striking accuracy', 'Grappling accuracy' and 'Sig. Strikes Landed', 'Sig. Strikes Attempted', 'Takedowns Landed', 'Takedowns Attempted'. Let's drop these columns : 

```python
df.drop(columns={'wins by : KO/TKO %', 'wins by : Decision %', 'wins by Submission %',
                'Striking accuracy', 'Grappling accuracy'}, axis=1, inplace=True)
```

some columns have decimal values instead of integer values, let's fix that :

```python
# Converting columns type from float to int

df[['Age', 'Sig.str by target : HEAD', 'Sig.str by target : BODY', 'Sig.str by target : LEG',
    'Fight Win Streak', 'Wins by Knockout', 'Wins by Submission', 'Sig. Str. Absorbed per win',
        'Sig. Str. Defense', 'Takedown Defense', 'Sig. Strikes Landed', 'Sig. Strikes Attempted',
            'Takedowns Landed', 'Takedowns Attempted', 'Wins by Decision', 'First Round Finishes',
                 'Title Defenses']] = df[['Age', 'Sig.str by target : HEAD', 'Sig.str by target : BODY',
    'Sig.str by target : LEG', 'Fight Win Streak', 'Wins by Knockout',
        'Wins by Submission', 'Sig. Str. Absorbed per win', 'Sig. Str. Defense',
            'Takedown Defense', 'Sig. Strikes Landed', 'Sig. Strikes Attempted',
                'Takedowns Landed','Takedowns Attempted', 'Wins by Decision', 'First Round Finishes', 'Title Defenses']].fillna(0).astype(int)
```

```python
# checking other columns
df[['Sig.str : Standing', 'Sig.str : Clinch', 'Sig.str : Ground']].head(2)
```

we need to get rid of the % + converting the columns type to integer :

```python
df['Sig.str : Standing'] = df['Sig.str : Standing'].fillna('0').apply(lambda x : x.split(' ')[0]).astype(int)
df['Sig.str : Clinch'] = df['Sig.str : Clinch'].fillna('0').apply(lambda x : x.split(' ')[0]).astype(int)
df['Sig.str : Ground'] = df['Sig.str : Ground'].fillna('0').apply(lambda x : x.split(' ')[0]).astype(int)
```
